## Condition Based Filtering

```python
mask = titanic_data["Fare"] > 100
```

```python
fare_100  = titanic_data[mask]
```

```python
fare_100.shape
```

```python
fare_100.head(10)
```

```python
female_fare_100 = fare_100[fare_100.Sex == "female"]
```

```python
female_fare_100.shape
## Why nearly 65% of highest paying customer are female in 1912 ?
```

```python
female_fare_100 = titanic_data[
    (titanic_data.Fare > 100) & (titanic_data.Sex == "female")
    ]
female_fare_100.head()
```

```python
# Female survival Rate in Titanic

survived_female = titanic_data[
    (titanic_data['Sex'] == "female") & (titanic_data.Survived == 1)
]
survived_female.shape
```

Out of 891 passengers , 233 female passengers survived in the crash.

```python
survived_highclass = titanic_data[
    (titanic_data['Pclass'] == 1) & (titanic_data.Survived == 1)
]
survived_highclass.shape
```
