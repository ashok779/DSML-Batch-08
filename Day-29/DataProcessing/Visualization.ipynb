## 1. Bar Chart

```python
titanic_df["Survived"].value_counts().plot(
    kind = "bar",
    title = "Count of Survived Passengers",
    figsize = (4,3),
    legend = True,
    xlabel = "Survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    yticks = list(range(0,600,70)),
    colormap = "Paired"
)
```

```python
sns.countplot(
    data = titanic_df,
    x = "Survived", 
    hue = "Sex",
)
```

```python
titanic_df[titanic_df["Sex"] == "male"]["Survived"].value_counts().plot(
    kind = "bar",
    title = "Count of Male Survived Passengers",
    figsize = (4,3),
    legend = True,
    xlabel = "Survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    yticks = list(range(0,600,70)),
    colormap = "Paired"
)
```

```python
group_by_gender = titanic_df[["Sex","Survived"]].groupby(by = "Sex")

group_by_gender["Survived"].value_counts().plot(
    kind = "bar",
    title = "Count of Male Survived Passengers",
    figsize = (4,3),
    legend = True,
    xlabel = "Survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    yticks = list(range(0,600,70)),
    colormap = "Paired"
)
```

```python
pd.crosstab(
    titanic_df.Survived,
    titanic_df.Sex
).plot(
    kind = "bar",
    title = "Count of Survived Passengers",
    figsize = (4,3),
    legend = True,
    xlabel = "Survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    yticks = list(range(0,600,70)),
    colormap = "tab20"
)
```

## 2. Pie Chart

```python
group_by_embarked = titanic_df.groupby(by = "Embarked")
group_by_embarked.count()
```

```python
group_by_embarked.count().plot(
    kind = "pie", y = "Survived",
    title = "Count of Survived Passengers",
    figsize = (4,3),
    legend = True,
    colormap = "tab20",
    autopct = '%1.1f%%',
    startangle = 90
)
```

```python
pd.crosstab(
    titanic_df.Survived,
    titanic_df.Sex
).plot(
    kind = "pie", subplots = True,
    title = "Count of Survived Passengers",
    figsize = (5,4),
    legend = True,
    colormap = "tab20"
)
```

## 3. Histogram

```python
titanic_df.Age.plot(
    kind = "hist", 
    grid = True,
    bins = 20
)
```

```python
titanic_df[["Age","Sex"]].plot(
    by = "Sex",
    kind = "hist", 
    grid = True,
    bins = 20
)
```

```python
titanic_df[["Age","Fare"]].plot(
    kind = "hist", 
    grid = True,
    bins = 20,
)
```

## 4. Line Graph

```python
titanic_df[["Fare", "Age"]].loc[:200].plot(
    kind = "line",
    figsize = (15,4),
    xticks = list(range(0,200,10)),
    xlabel = "Row",
    ylabel = "Value"

)
```
