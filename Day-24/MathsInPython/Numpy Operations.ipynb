## 1. Elementwise Operations


```python
import numpy as np
```

```python
a = np.array(
    [
        [4,5,6],
        [1,2,4]
    ]
)
b = np.array([[8,6,0],[3,2,4]])
c = np.array([[8,4,0],[3,11,4]])
```

```python
a, b 
```

### Arithmetic Operations

```python
# a + b
np.add(a,b)
```

```python
# a - b
np.subtract(a,b)
```

```python
# a ** 2
np.pow(a, 2)
```

```python
# a % b
quo , rem = np.divmod(a,b)
quo
```

```python
rem
```

### Trignometric

```python
np.sin(a)
```

```python
cot = 1 / np.tan(a)
cot
```

### Rounding operations

```python
np.around(cot, 2)
```

```python
np.floor(cot)
```

```python
np.ceil(cot)
```

```python
# Inverse of an matrix

sq = np.random.randint(0,10, (3,3))
a_det = np.linalg.det(sq)
```

```python
inv_sq = np.linalg.inv(sq)
```

```python
np.dot(sq, inv_sq).round()
```
