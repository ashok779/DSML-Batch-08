```python
import pickle
import numpy as np
import tensorflow as tf
```

```python
with open("iris_classifier.pkl", "rb") as file:
    objects = pickle.load(file)
```

```python
scaler = objects['scaler']
encoder = objects['encoder']
model = objects['model']
```

```python
model_tf = tf.keras.models.load_model("iris_classification.keras")
```

```python
# User Input

my_data = np.array([
    [12.6, 2.6, 8.5, 0.9],
    [6.1, 3.0, 4.9, 1.8]
])
```

```python
# Pre Processing

scaled_data = scaler.transform(my_data)
scaled_data
```

```python
# Predict

output = np.argmax(model.predict(scaled_data), axis = 1)
output
```

```python
cat = encoder.categories_[0]
values = [cat[res] for res in output]
values
```
